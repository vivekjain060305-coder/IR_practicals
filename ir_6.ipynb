```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
documents = [
    "Information Retrieval systems are used to search documents.",
    "Machine learning can be applied in various fields.",
    "Natural Language Processing deals with language data."
]
```

```python
query = "Information search in documents using algorithms"
```

```python
def cosine_sim(doc1, doc2):
    vectorizer = CountVectorizer().fit_transform([doc1, doc2])
    cosine_sim_matrix = cosine_similarity(vectorizer)
    return cosine_sim_matrix[0, 1]
```

```python
def jaccard_sim(doc1, doc2):
    set1 = set(doc1.split())
    set2 = set(doc2.split())
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))
    return intersection / union
```

```python
def rank_documents(documents, query):
    cosine_scores = []
    jaccard_scores = []

    for doc in documents:
        cosine_scores.append(cosine_sim(query, doc))
        jaccard_scores.append(jaccard_sim(query, doc))

    cosine_ranking = np.argsort(cosine_scores)[::-1]
    jaccard_ranking = np.argsort(jaccard_scores)[::-1]

    return cosine_scores, jaccard_scores, cosine_ranking, jaccard_ranking

cosine_scores, jaccard_scores, cosine_ranking, jaccard_ranking = rank_documents(documents, query)

for i, doc in enumerate(documents):
    print(f"Document {i+1}: {doc}")
    print(f"Cosine Similarity: {cosine_scores[i]:.4f}, Jaccard Similarity: {jaccard_scores[i]:.4f}\n")

print("\nRankings based on Cosine Similarity:")
for idx in cosine_ranking:
    print(f"Document {idx+1}: {documents[idx]}")

print("\nRankings based on Jaccard Similarity:")
for idx in jaccard_ranking:
    print(f"Document {idx+1}: {documents[idx]}")
```
