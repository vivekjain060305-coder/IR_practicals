AIM: implement boolean query processing using AND OR NOT operators to retrive document based on user queries..

```python
documents = {
    1: "information retrieval is fun",
    2: "boolean retrieval uses and or not",
    3: "information systems and databases",
    4: "retrieval of information using boolean queries"
}
```

```python
from collections import defaultdict

inverted_index = defaultdict(set)

for doc_id, text in documents.items():
    words = text.lower().split()
    for word in words:
        inverted_index[word].add(doc_id)

print("Inverted Index:")
for term, docs in inverted_index.items():
    print(term, ":", docs)
```

```python
def boolean_and(term1, term2):
    return inverted_index[term1] & inverted_index[term2]

def boolean_or(term1, term2):
    return inverted_index[term1] | inverted_index[term2]

def boolean_not(term):
    all_docs = set(documents.keys())
    return all_docs - inverted_index[term]
```

```python
def process_query(query):
    tokens = query.lower().split()

    if "and" in tokens:
        term1 = tokens[0]
        term2 = tokens[2]
        result = boolean_and(term1, term2)

    elif "or" in tokens:
        term1 = tokens[0]
        term2 = tokens[2]
        result = boolean_or(term1, term2)

    elif "not" in tokens:
        term = tokens[1]
        result = boolean_not(term)

    else:
        result = inverted_index[tokens[0]]

    return result
```

```python
query = input("Enter Boolean Query (AND / OR / NOT): ")
result_docs = process_query(query)

print("\nDocuments Retrieved:")
if result_docs:
    for doc_id in result_docs:
        print(f"Doc {doc_id}:", documents[doc_id])
else:
    print("No documents found")
```
