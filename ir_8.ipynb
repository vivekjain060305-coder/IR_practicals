```python
import pandas as pd

df = pd.read_csv("diabetes_prediction_dataset.csv")
df.head()
```

```python
df['predicted_diabetes'] = df['HbA1c_level'].apply(lambda x: 1 if x > 6 else 0)

df[['HbA1c_level','diabetes','predicted_diabetes']].head()
```

```python
TP = len(df[(df['predicted_diabetes']==1) & (df['diabetes']==1)])
TN = len(df[(df['predicted_diabetes']==0) & (df['diabetes']==0)])
FP = len(df[(df['predicted_diabetes']==1) & (df['diabetes']==0)])
FN = len(df[(df['predicted_diabetes']==0) & (df['diabetes']==1)])

print("True Positive:", TP)
print("True Negative:",TN)
print("False Positive:", FP)
print("False Negative:", FN)
```

```python
accuracy = (TP + TN) / (TP + TN + FP + FN)

print("Accuracy:", accuracy)
```

```python
print("Accuracy Percentage:", round(accuracy*100,2), "%")
```

```python
precision = TP / (TP + FP)
print("Precision:", precision)
```

```python
recall = TP / (TP + FN)
print("Recall:", recall)
```

```python
f1_score = 2 * (precision * recall) / (precision + recall)
print("F1 Score:", f1_score)
```

```python
print("\nEvaluation Metrics")

print("Precision :", round(precision,3))
print("Recall    :", round(recall,3))
print("F1 Score  :", round(f1_score,3))
```
