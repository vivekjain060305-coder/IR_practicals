Aim: To implement the TF–IDF weighting technique for a given document corpus and generate a TF–IDF matrix for effective document representation.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

documents = [
    "Information Retrieval systems are used to search documents.",
    "Machine learning can be applied in various fields.",
    "Natural Language Processing deals with language data."
]

vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(documents)

feature_names = vectorizer.get_feature_names_out()

tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names)

print("TF-IDF Matrix:\n")
print(tfidf_df)
```
