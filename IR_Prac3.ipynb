```python
#Vivek Jain
#CS23094
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
#Vivek Jain
#CS23094
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')
```

```python
#Vivek Jain
#CS23094
text = """
Information Retrieval is the process of obtaining relevant information
from a large collection of unstructured documents. Text preprocessing
plays a crucial role in improving retrieval performance.
"""
print("Original Text:\n", text)
```

```python
#Vivek Jain
#CS23094
tokens = word_tokenize(text)
print("Tokens:\n", tokens)
```

```python
#Vivek Jain
#CS23094
stop_words = set(stopwords.words('english'))

filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
print("After Stop-word Removal:\n", filtered_tokens)
```

```python
#Vivek Jain
#CS23094
ps = PorterStemmer()
stemmed_tokens = [ps.stem(word) for word in filtered_tokens]
print("After Stemming:\n", stemmed_tokens)
```

```python
#Vivek Jain
#CS23094
print("Final Preprocessed Text:\n", stemmed_tokens)
```
