Aim: Representation of Text document in Vector Space Model and computing similarity between two document.

```python
import math
docs = {
    "d1": {"apple", "banana", "orange"},
    "d2": {"banana", "grape", "kiwi"},
    "d3": {"apple", "kiwi", "peach"},
    "d4": {"orange", "grape", "mango"}
}
vocab = sorted(set().union(*docs.values()))
def vectorize(doc):
    return [1 if term in docs[doc] else 0 for term in vocab]
def cosine_similarity(doc1, doc2):
    v1 = vectorize(doc1)
    v2 = vectorize(doc2)
    dot_product = sum(a*b for a, b in zip(v1, v2))
    mag1 = math.sqrt(sum(a*a for a in v1))
    mag2 = math.sqrt(sum(b*b for b in v2))
    if mag1 == 0 or mag2 == 0:
        return 0
    return dot_product / (mag1 * mag2)
print("Vocabulary :", vocab)
print("Vector of d1 :", vectorize("d1"))
print("Vector of d2 :", vectorize("d2"))
print("Cosine Similarity between d1 and d2 :", cosine_similarity("d1", "d2"))
```
