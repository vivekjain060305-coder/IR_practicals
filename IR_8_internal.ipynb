```python
import pandas as pd
data = pd.read_csv("/content/drive/MyDrive/ML_lab/Mall_Customers.csv")
print("First 5 Records of Dataset:")
print(data.head())
data['Relevant'] = data['Spending Score (1-100)'] > 50
data['Retrieved'] = data['Annual Income (k$)'] > 50
TP = len(data[(data['Relevant'] == True) & (data['Retrieved'] == True)])
FP = len(data[(data['Relevant'] == False) & (data['Retrieved'] == True)])
FN = len(data[(data['Relevant'] == True) & (data['Retrieved'] == False)])
TN = len(data[(data['Relevant'] == False) & (data['Retrieved'] == False)])
precision = TP / (TP + FP) if (TP + FP) != 0 else 0
recall = TP / (TP + FN) if (TP + FN) != 0 else 0
f_measure = (2 * precision * recall) / (precision + recall) if (precision + recall) != 0 else 0
accuracy = (TP + TN) / (TP + TN + FP + FN)
print("\nIR Evaluation Metrics")
print("-------------------------")
print("True Positive (TP):", TP)
print("False Positive (FP):", FP)
print("False Negative (FN):", FN)
print("True Negative (TN):", TN)
print("\nPrecision:", round(precision,3))
print("Recall:", round(recall,3))
print("F-Measure:", round(f_measure,3))
print("Accuracy:", round(accuracy,3))
```
